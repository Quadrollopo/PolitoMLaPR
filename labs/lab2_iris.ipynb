# Lab 2 - Visualization on Iris Dataset
## Imports

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from warnings import warn
from tiblib.visualization import pair_plot
```

## Loading

```python
iris = pd.read_csv('../datasets/iris.csv',header=None).to_numpy()
X = iris[:,:-1]
y = iris[:,-1]
X_centered = X - np.mean(X, axis=0)
```

## Visualization

```python
columns = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']

pair_plot(X,y, columns=columns) # Normal plot
pair_plot(X_centered,y, columns=columns) # Centered plot
```
