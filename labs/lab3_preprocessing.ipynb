# Lab 3 - PCA on Iris Dataset
## Imports

```python
import pandas as pd
from tiblib.visualization import plot_2d
from tiblib.preprocessing import LDA, PCA
```

## Loading

```python
iris = pd.read_csv('../datasets/iris.csv',header=None)
X = iris.iloc[:,:-1].to_numpy()
y = iris.iloc[:,-1].to_numpy()
```

## PCA

```python
pca = PCA(n_dims=2)
X_reduced = pca.fit_transform(X,y)
plot_2d(X_reduced,y)
```

## LDA

```python
lda = LDA(n_dims=2)
X_reduced = lda.fit_transform(X, y)
plot_2d(X_reduced, y)
```
